# Preprocesamiento

Un directivo del bano está preocupado porque cada vez más clientes abandonan sus servicios de tarjetas de crédito.Se pide predecir quién se va a abandonar el banco, para así ofrecerle mejor servicios y hacer que el cliente cambie de opinión.

El conjunto de datos [Credit Card customers](https://www.kaggle.com/datasets/sakshigoyal7/credit-card-customers?select=BankChurners.csv) fue obtenida de Kaggle.

Este conjunto de datos consta de 10000 clientes con los siguientes campos:

*   **CLIENTNUM:** Identificación de cliente.
*   **Attrition_Flag:**Actividad del cliente.
*   **Customer_Age:** Edad delcliente.
*   **Gender:** F=Mujer, M=Masculino.
*   **Dependent_count**
*   **Education_Level:**Grado académido del titular de la cuenta.
*   **Marital_Status:**Casado, Soltero, Divorciado, Desconocido.
*   **Income_Category:**Categoría de ingreso anual del titular de la cuenta (<\$40K, \$40K-\$60K, \$60K-\$80K,  \$80K-\$120K, >\$120K, Desconocido)
*   **Card_Category:**Tipo de Tarjeta (Blue, Silver, Gold, Platinum).
*   **Months_on_book:**Periodo de relación con el banco.
*   **Total_Relationship_Count:**Número total de producto en poder del cliente.
*   **Months_Inactive_12_mon:**N° de meses inactivos en los últimos 12 meses.
*   **Contacts_Count_12_mon:**N° de contactos en los últimos 12 meses.
*   **Credit_Limit:**Limite de crédito en la tarjeta de crédito.
*   **Total_Revolving_Bal:**Saldo rotativo de la tarjeta de crédito.
*   **Avg_Open_To_Buy:**Línea de crédito abierta para comprar (promedio de los últimos 12 meses).
*   **Total_Amt_Chng_Q4_Q1:**Variación del importe de las transacciones(Q4 sobre Q1).
*   **Total_Trans_Amt:**Importe total de la transacción(últimos 12 meses)
*   **Total_Trans_Ct:**Recuento total de transacciones(últimos 12 meses).
*   **Total_Ct_Chng_Q4_Q1:**Cambio en el recuento de transacciones(Q4 sobre Q1).
*   **Avg_Utilization_Ratio:**Radio de uso medio de la tarjeta.

## Pasar las variables categoricas a numericas

Attrition_Flag
Gender
Education_Level
Marital_Status
Card_Category

```python
data_num = data.copy()
```

```python
categoricas = {
    "Attrition_Flag": {"Existing Customer":1, "Attrited Customer":0},
    "Gender": {"F":0, "M":1},
    "Education_Level": {"Uneducated": 0, "High School": 1, "Graduate": 2,
                        "College": 3, "Post-Graduate": 4, "Doctorate": 5,"Unknown": 6},
    "Marital_Status": {"Single": 0, "Married": 1, "Divorced": 3, "Unknown" :4},
    "Card_Category": {"Blue": 0, "Silver": 1, "Gold": 2, "Platinum": 3},
    "Income_Category": {"Less than $40K": 0, "$40K - $60K": 1 , "$60K - $80K": 2, 
                        "$80K - $120K": 3, "$120K +":4, "Unknown": 5}
}
data_num.replace(categoricas, inplace=True)
```

```python
data_num.head()
```

```python
# eliminando los valores desconocidos en cada variable
data_num.drop(data_num[data_num["Education_Level"]== 6].index, inplace=True )
data_num.drop(data_num[data_num["Marital_Status"]== 4].index, inplace=True )
data_num.drop(data_num[data_num["Income_Category"]== 5].index, inplace=True )
```

```python
data = data.loc[data_num.index]
data = data.reset_index(drop=True)
data.head()
```

```python
data_num = data_num.reset_index(drop=True)
data_num.head()
```

```python
# exportar data
data_num.to_csv("credit-card-customers-numerica.csv")
data.to_csv("credit-card-customers.csv")
```
